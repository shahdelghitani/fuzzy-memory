```python
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
```

```python
data = pd.read_csv(r'c:\Users\skynet\AppData\Local\Temp\Rar$DRa13716.39352\covid_19_data.csv')
data
```

```python


# تحويل عمود 'ObservationDate' إلى نوع datetime
data['ObservationDate'] = pd.to_datetime(data['ObservationDate'], format='%m/%d/%Y')

# تجهيز البيانات للعرض
data_grouped = data.groupby(['ObservationDate', 'Country/Region']).sum().reset_index()

# بناء التطبيق
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("COVID-19 Dashboard"),
    
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in data['Country/Region'].unique()],
        value='Mainland China',  # قيمة افتراضية
        style={'width': '50%'}
    ),
    
    dcc.Graph(id='cases-graph'),
    dcc.Graph(id='deaths-graph'),
    dcc.Graph(id='recovered-graph')
])

@app.callback(
    [Output('cases-graph', 'figure'),
     Output('deaths-graph', 'figure'),
     Output('recovered-graph', 'figure')],
    [Input('country-dropdown', 'value')]
)
def update_graphs(selected_country):
    df_country = data_grouped[data_grouped['Country/Region'] == selected_country]
    
    cases_fig = px.line(df_country, x='ObservationDate', y='Confirmed', title=f'Confirmed Cases in {selected_country}')
    deaths_fig = px.line(df_country, x='ObservationDate', y='Deaths', title=f'Deaths in {selected_country}')
    recovered_fig = px.line(df_country, x='ObservationDate', y='Recovered', title=f'Recovered Cases in {selected_country}')
    
    return cases_fig, deaths_fig, recovered_fig

if __name__ == '__main__':
    app.run_server(debug=True)

```
